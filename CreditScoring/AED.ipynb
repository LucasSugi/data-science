```python
#Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
#Leitura da base
df = pd.read_excel('Base4.xls')
df.columns = ['idade','educacao','t_emprego','t_endereco','renda','divida','divida_cc','outras_div','default']

#Trocando a classe de 0(1) para 1(0)
df['default'] = df['default'].map(lambda x: 0 if x == 1 else 1)

#Verificando os tipos lidos
print(df.dtypes)

#Transformando educacao e default como variaveis categoricas
df = df.astype({'default':'str','educacao':'str'})

#Mostrando parte do dataset
df.head()
```

```python
#Verificando missing values
print(df.isnull().sum(axis=0))

#Análise descritiva
print(df.describe(),'\n')

#Aqui ja podemos identificar pelo max e terceiro quartil que existem valores muito grandes em renda, divida e divida_cc
#quando comparados ao resto do grupo (possíveis outliers)
```

```python
#Verificando balanceamento do dataset para as variaveis categoricas
print(df.groupby('default').size(),'\n')
print(df.groupby('educacao').size())

#Note que ambas as variaveis possuem um desbalanceamento
```

```python
#Intervalando a variavel educacao dado o seu desbalanceamento
df['educacao'] = df['educacao'].map(lambda x: '3' if(x == '4' or x == '5') else x)
df.groupby('educacao').size()
```

```python
#Boxplot das variaveis numericas
plt.figure(figsize=(15,15))
plt.subplots_adjust(wspace=0.7)
plt.subplot(331)
ax = sns.boxplot(data=df,x=df['default'],y=df['idade'])
ax.set_xlabel('default',fontsize=15)
ax.set_ylabel('idade',fontsize=15)
ax.tick_params(labelsize=15)

plt.subplot(332)
ax = sns.boxplot(data=df,x=df['default'],y=df['t_emprego'])
ax.set_xlabel('default',fontsize=15)
ax.set_ylabel('t_emprego',fontsize=15)
ax.tick_params(labelsize=15)

plt.subplot(333)
ax = sns.boxplot(data=df,x=df['default'],y=df['t_endereco'])
ax.set_xlabel('default',fontsize=15)
ax.set_ylabel('t_endereco',fontsize=15)
ax.tick_params(labelsize=15)

plt.subplot(334)
ax = sns.boxplot(data=df,x=df['default'],y=df['renda'])
ax.set_xlabel('default',fontsize=15)
ax.set_ylabel('renda',fontsize=15)
ax.tick_params(labelsize=15)

plt.subplot(335)
ax = sns.boxplot(data=df,x=df['default'],y=df['divida'])
ax.set_xlabel('default',fontsize=15)
ax.set_ylabel('divida',fontsize=15)
ax.tick_params(labelsize=15)

plt.subplot(336)
ax = sns.boxplot(data=df,x=df['default'],y=df['divida_cc'])
ax.set_xlabel('default',fontsize=15)
ax.set_ylabel('divida_cc',fontsize=15)
ax.tick_params(labelsize=15)

plt.subplot(337)
ax = sns.boxplot(data=df,x=df['default'],y=df['outras_div'])
ax.set_xlabel('default',fontsize=15)
ax.set_ylabel('outras_div',fontsize=15)
ax.tick_params(labelsize=15)
plt.show()
```

```python
#Pelo boxplot anterior aplicaremos o logaritmo em outras_div, divida_cc e renda para distribuir melhor os dados
df['renda'] = df['renda'].map(lambda x: np.log(x))
df['outras_div'] = df['outras_div'].map(lambda x: np.log(x))
df['divida_cc'] = df['divida_cc'].map(lambda x: np.log(x))

#Novos boxplot's com variaveis transformadas
plt.figure(figsize=(15,5))
plt.subplots_adjust(wspace=0.7)
plt.subplot(131)
ax = sns.boxplot(data=df,x=df['default'],y=df['renda'])
ax.set_xlabel('default',fontsize=15)
ax.set_ylabel('renda',fontsize=15)
ax.tick_params(labelsize=15)

plt.subplot(132)
ax = sns.boxplot(data=df,x=df['default'],y=df['outras_div'])
ax.set_xlabel('default',fontsize=15)
ax.set_ylabel('ouras_div',fontsize=15)
ax.tick_params(labelsize=15)

plt.subplot(133)
ax = sns.boxplot(data=df,x=df['default'],y=df['divida_cc'])
ax.set_xlabel('default',fontsize=15)
ax.set_ylabel('divida_cc',fontsize=15)
ax.tick_params(labelsize=15)
plt.show()

#Note pelos boxplot's que como há um espalhamento maior dos valores
#os outliers acabaram ficando próximos de todos, amenizando pontos
#fora do padrão
```

```python
#ScatterPlot com gráfico de densidades
sns.set_context("paper", rc={"axes.labelsize":20})
ax = sns.pairplot(data=df,hue='default',vars=['idade','t_emprego','t_endereco','renda','divida','divida_cc','outras_div'],diag_kind='kde')
plt.show()
```

```python
#Matriz de correlação via heatmap
corr = df.iloc[:,[0,2,3,4,5,6,7]].corr()
plt.figure(figsize=(12,12))
ax = sns.heatmap(data=corr,vmin=0,vmax=1,linewidths=.5)
ax.tick_params(labelsize=13)
plt.show()

#Pelo scatterPlot e heatmap notamos que as seguintes variaveis são mais
#correlacionadas e devem ser verificadas quanto a colineariedade e interação

# renda:t_emprego
# outras_div:renda
# outras_div:divida_cc
# divida_cc:renda
# t_endereco:idade

#Divida nao precisa ser avaliada quanto a interação pois é formada pelas variaveis outras_div, divida_cc e renda
```

```python
#Iremos criar uma nova variavel que ira condensar outras_div e divida_cc
#acreditando que ambas fornecem um poder descritivo maior juntas
df['div'] = df['outras_div'] + df['divida_cc']
plt.figure(figsize=(7,7))
ax = sns.boxplot(data=df,x='default',y='div')
ax.set_xlabel('default',fontsize=15)
ax.set_ylabel('div',fontsize=15)
ax.tick_params(labelsize=15)
plt.show()

#Tirando a variavel idade e educacao, as outras aparentam discretizar bem o problema.
#Para essas duas não se sabe muito bem sobre seu poder e aparentemente não será significante.
#Porém, deve-se considerar analisar se há interações delas com outras variáveis.
```
